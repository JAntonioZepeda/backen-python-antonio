```python
def es_entero(valor):
    try:
        # Intenta convertir el valor a un entero
        int_valor = int(valor)
        # Si la conversión es exitosa, entonces el valor es un entero
        return True
    except ValueError:
        # Si la conversión genera una excepción ValueError, el valor no es un entero
        return False   
while True:
    suma_fila=0
    suma_set=0
    suma_diccionario=0 
    print("Opciones: \n")
    print("1. Suma")
    print("2. Resta")
    print("3. Multiplicación")
    print("4. División")
    print("5. Potencia")
    print("6. Salir") 
    opcion = input("Elige una opción (1/2/3/4/5/6): ")
    if opcion=='6':
        print("Saliendo de la calculadora")
        break
    if opcion not in('1','2','3','4','5','6'):
        print("Opcion no valida")
        continue
    lista=[]
    my_set=set()
    mi_diccionario = {}
    print("Llenando lista: ")
    i=0
    while i < 5:
        valor= input(f"Ingresa el primer valor de la fila {i+1}:")
        if es_entero(valor):
            lista.append(int(valor))
            i+=1    
        else: 
            print("ERROR! Entrada invalida.")   
    a=0      
    print("Llenando Set: ")
    while a < 5:
        valor2= input(f"Ingresa el primer valor de set {a+1}:")
        if es_entero(valor2):
            my_set.add(int(valor2))
            a+=1
        else:
            print("ERROR! Entrada invalida.")      
    b=0        
    print("Llenando Diccionario: ")
    while b < 5:
        valor3= input(f"Ingresa el primer valor del diccionario {b+1}:")
        if es_entero(valor3):
            mi_diccionario[b]=int(valor3)
            b+=1
        else:
            print("ERROR! Entrada invalida.")
            
    if opcion=='1':
        suma_fila=0
        suma_set=0
        suma_diccionario=0 
        for elemento in lista:
            suma_fila+=elemento
        print("lista:", lista," ",type(lista),"La suma de la fila es: ",suma_fila) 
        for elemento1 in my_set:
            suma_set+=elemento1
        print("Set:",my_set," ", type(my_set),"La suma del set es: ",suma_set)
        for elemento2 in mi_diccionario.values():
            suma_diccionario+=elemento2
        print("Diccionario:",mi_diccionario," ",type(mi_diccionario),"La suma del diccionario es: ",suma_diccionario)
      
    if opcion=='2':
        suma_fila=2
        suma_set=2
        suma_diccionario=2 
        for elemento in lista:
            suma_fila-=elemento
        print("lista:", lista," ",type(lista),"La resta de la fila es: ",suma_fila) 
        for elemento1 in my_set:
            suma_set-=elemento1
        print("Set:",my_set," ", type(my_set),"La resta del set es: ",suma_set)
        for elemento2 in mi_diccionario.values():
            suma_diccionario-=elemento2
        print("Diccionario:",mi_diccionario," ",type(mi_diccionario),"La resta del diccionario es: ",suma_diccionario)
    if opcion=='3':
        suma_fila=1
        suma_set=1
        suma_diccionario=1
        for elemento in lista:
            suma_fila*=elemento
        print("lista:", lista," ",type(lista),"La multiplicacion de la fila es: ",suma_fila) 
        for elemento1 in my_set:
            suma_set*=elemento1
        print("Set:",my_set," ", type(my_set),"La multiplicacion del set es: ",suma_set)
        for elemento2 in mi_diccionario.values():
            suma_diccionario*=elemento2
        print("Diccionario:",mi_diccionario," ",type(mi_diccionario),"La multiplicacion del diccionario es: ",suma_diccionario)
    if opcion=='4':
        resultado = 1
        for valor in lista:
            if valor!= 0:
                resultado /= valor
            else:
                print("tamaño inválido en la entrada:división por 0 ")              
        print("Lista:",lista," ",type(lista),"la división de la fila es: ",resultado)
        resultado2 = 1
        for valor1 in my_set:
            if valor1!= 0:
                resultado2 /= valor1
            else:
                print("tamaño inválido en la entrada:división por 0 ")              
        print("Lista:",my_set," ",type(my_set),"la división del set es: ",resultado2)
        resultado3 = 1
        for valor2 in mi_diccionario.values():
            if valor2!= 0:
                resultado3 /= valor2
            else:
                print("tamaño inválido en la entrada:división por 0 ")
                
        print("Lista:",mi_diccionario," ",type(mi_diccionario),"la división del diccionario es: ",resultado3)
    if opcion=='5': 
        resultado = 1
        for valor in lista:
                resultado *= valor
        print("Lista:",lista," ",type(lista),"la Potencia de la fila es: ",resultado)
        resultado2 = 1
        for valor1 in my_set:
                resultado2 *= valor1         
        print("Lista:",my_set," ",type(my_set),"la Potencia del set es: ",resultado2)
        resultado3 = 1
        for valor2 in mi_diccionario.values():        
                resultado3 *= valor2             
        print("Lista:",mi_diccionario," ",type(mi_diccionario),"la Potencia del diccionario es: ",resultado3)
        
```
